```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Name' : ['Harshita', 'Pooja', 'Jahnvi', 'Deepak', 'Shivkant', 'Anushka'],
    'Age' : [20, 19, 19, 25, 23, 20],
    'Department' : ['CS', 'IT', 'CS', 'CS', 'IT', 'IT']
}

data
```
